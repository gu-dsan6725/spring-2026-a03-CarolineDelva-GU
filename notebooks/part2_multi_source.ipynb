```python
import sys
import os

sys.path.append(os.path.abspath(".."))

from src.rag_pipeline import answer_query

questions = [
    "What is the total sales recorded?",
    "Describe the main features of the products",
    "Find mentions of battery life in reviews"
]

results = []

for q in questions:
    ans = answer_query(q)
    print(ans)
    results.append((q, ans))
```

```python
with open("part2_results.txt", "w") as f:

    for q, ans in results:
        f.write(f"QUESTION: {q}\n")
        f.write(ans)
        f.write("\n\n")
```
